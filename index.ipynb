```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 

df = pd.read_csv('./train.csv')
df.head()
print('********Age stats:')
#Age
print('Mean Age: {}'.format(df['age'].mean()))
print('Median Age: {}'.format(df['age'].median()))
print('Age quantile (90%): {}'.format(df['age'].quantile(.9)))
print('Age quantile (50%): {}'.format(df['age'].quantile(.5)))

#crime
print('********Crime stats:')
print('Mean crime: {}'.format(df['crim'].mean()))
print('Median crime: {}'.format(df['crim'].median()))
print('crime quantile (90%): {}'.format(df['crim'].quantile(.9)))
print('crime quantile (50%): {}'.format(df['crim'].quantile(.5)))


#Tax
print('********Tax stats:')
print('Mean Tax: {}'.format(df['tax'].mean()))
print('Median Tax: {}'.format(df['tax'].median()))
print('Tax quantile (90%): {}'.format(df['tax'].quantile(.9)))
print('Tax quantile (50%): {}'.format(df['tax'].quantile(.5)))




# #properties next to Charles River
df_nxt_to_charles_river = df.loc[df['chas'] == 1]
df_nxt_to_charles_river.head()
#properties Away from Charles river
df_away_charles_river = df.loc[df['chas'] == 0]
df_away_charles_river.head()


# #Areas with per capita Crime > 0.26
df_crime_gt_026 = df.loc[df['crim'] > 0.26]
df_crime_gt_026.head()
#Areas with per capita Crime <= 0.26
df_crime_le_026 = df.loc[df['crim'] <= 0.26]
df_crime_le_026.head()

df_crim_river_medianValue = df.loc[: , ['crim','chas','medv']]
df_crim_river_medianValue.plot()

df_crim_river_medianValue.plot.box()

# Define a new figure with matplotlib's .plot() function. Set the size of figure space
figure = plt.figure(figsize=(10,4))

# Add a subplot to the figure - a new axes
ax = figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = figure.add_subplot(122)

# Generate a line plot on first axes
ax.plot(df_nxt_to_charles_river.medv, df_nxt_to_charles_river.crim, color='lightblue', linewidth=3, linestyle = '-.')

# Draw a scatter plot on 2nd axes
ax2.plot(df_away_charles_river.medv, df_away_charles_river.crim, color='red', linewidth=3, linestyle = '-.')

ax.set_title('Properties Next to Charles River')
ax.set_xlabel('Property Median Value')
ax.set_ylabel('Crime Rate')

ax2.set_title('Properties Away From Charles River')
ax2.set_xlabel('Property Median Value')
ax2.set_ylabel('Crime Rate')

# Show the plot
plt.show()



#*******************
# Creates blank canvas
figure = plt.figure()

# Add new axes to the figure with absolute positions
ax1 = figure.add_axes([.9, .9, .9, .9]) # main axes
ax2 = figure.add_axes([.2, .5, .4, .3]) # inset axes

# Larger Figure Axes 1
ax1.plot(df_crime_gt_026.rm,df_crime_gt_026.medv, color = 'blue', linestyle = '-.')
ax1.set_xlabel('No. of rooms')
ax1.set_ylabel('Property Median Value')
ax1.set_title('Crime Rate > 0.26 per capita')

# Insert Figure Axes 2
ax2.plot(df_crime_le_026.rm,df_crime_le_026.medv, color = 'green', linestyle = '--')
ax2.set_xlabel('No. of rooms')
ax2.set_ylabel('No. of rooms')
ax2.set_title('Crime Rate <= 0.26 per capita')

plt.show()
```

# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
